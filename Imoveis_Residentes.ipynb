# **Relatório de Análise III**

```python
import pandas as pd
```

```python
dados = pd.read_csv('/content/drive/MyDrive/Colab Notebooks/Projetos Python/Curso Pandas/aluguel.csv', sep = ';')
```

```python
dados.head(10)
```

```python
# Criando uma lista de itens não duplicados
list(dados['Tipo'].drop_duplicates())
```

```python
residencial = [
    'Quitinete',
    'Casa',
    'Apartamento',
    'Casa de Condomínio',
    'Casa de Vila'
]
residencial
```

```python
dados.head(10)
```

```python
selecao = dados['Tipo'].isin(residencial)
selecao
```

```python
#Listar no DtaFrame somente os imóveis residenciais ('Quitinete', 'Casa', 'Apartamento', 'Casa de Condomínio', 'Casa de Vila')
dados_residencial = dados[selecao]
dados_residencial
```

```python
list(dados_residencial['Tipo'].drop_duplicates())
```

```python
#Total de Registros Residenciais
dados_residencial.shape[0]
```

```python
#Total Geraal de Registros
dados.shape[0]
```

```python
#Reconstruir o índice
dados_residencial.index = range(dados_residencial.shape[0])
dados_residencial.index
```
